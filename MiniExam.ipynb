# เริ่มทำข้อสอบ

```python
fi_mg = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน') #เชื่อมตาราง flights และ airports
fi_mg
```

```python
fi_mg.dropna() #ทำการตรวจ missing
```

## 0.หารัฐที่มีจำนวนเครื่องบินขาเข้ามากที่สุดและน้อยที่สุด

```python
fi_mg.groupby('รัฐ').count()  ## ดูจำนวนเที่ยวบินของแต่ละรัฐ
```

```python
max = fi_mg.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินต้นทาง'],ascending=False)  ## Group column รัฐ ที่มีสนามบินต้นทางมากสุด
max.iloc[:1,3:4]
```

```python
min = fi_mg.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินต้นทาง'],ascending=False) ## Group column รัฐ ที่มีสนามบินต้นทางน้อยสุด
min.iloc[-1:,3:4]
```

## 3.สายการบินไหนมีเที่ยวบินมากที่สุดและน้อยที่สุด

```python
fi_mg.groupby('สายการบิน').count()
```

```python
aw = fi_mg.groupby('สายการบิน').count().sort_values(by=['วันในเดือน'],ascending=False)
aw
```

```python
max = aw.iloc[:1,:1]
max
```

```python
min = aw.iloc[15:16,:1]
min
```
